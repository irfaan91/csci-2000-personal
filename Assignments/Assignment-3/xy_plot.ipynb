```python
#Irfaan Ali
#100400606

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#Importing data from file
f = open('xy_points.dat', mode="r")
line1 = []
line2 = []
line = f.readline()
for word in line.split():
    line1.append(word)
line = f.readline()
for word in line.split():
    line2.append(word)
f.close()

#Converting coordinates from string to float
x=[]
y=[]
for num in range(len(line1)):
    x.append(float(line1[num]))
for num in range(len(line2)):
    y.append(float(line2[num]))

    
#Saving coordinates that fall under the condition
# x < -2 and −5 ≤ y ≤ 0
bx = []
by = []
for i in range(len(x)):
    if ((x[i] < -2) and (y[i] >= -5) and (y[i]<=0)):
        bx.append(x[i])
        by.append(y[i])

#Saving coordinates that result in a radius of 5 or less
r = 5.0
gx = []
gy = []
for i in range(len(x)):
    rhs = (((x[i])**2) + ((y[i]-5)**2))**(1/2)
    if ((rhs <= r)):
        gx.append(x[i])
        gy.append(y[i])
        
        
        
plt.figure(figsize=(15,15), dpi=80)

#First plot, all points, red dots
plt.plot(x, y, 'ro')

#Second plot, points are between x < -2 and −5 ≤ y ≤ 0, blue dots
plt.plot(bx,by, 'bo')

#Third plot, points that fall in the circle at (0, 5) and
#has a radius of 5, green dots
plt.plot(gx,gy, 'go')

#Displays title of graph
plt.title('Highlighted points (100400606)')

#Limit on plot to [−10, 10] × [−10, 10]
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.show()


```
