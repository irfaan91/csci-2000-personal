```python
#Irfaan Ali
#100400606

#Used the triditional for loop to compute each sum
#Each function has the paramenter of 'n'

#Function definition for the summation of rho_pi
def rho_pi(n):
    ans = 0
    for k in range(1, n+1):
        ans += ((-1)**(k+1))/(2*k - 1)
    ans = 4*ans
    return ans

#Function definition for the summation of tau_pi
def tau_pi(n):
    ans = 0
    for k in range(1, n+1):
        ans += 6*(1/k**2)
    ans = ans**(1/2)
    return ans

#Function definition for the summation of mu_pi
def mu_pi(n):
    ans = 0
    for k in range(1, n+1):
        ans += 90*(1/k**4)
    ans = ans**(1/4)
    return ans

#Individual calls for each change in the variable 'n'
#pi approx for n = TEN
n = 10
ans = rho_pi(n)
print ("When n =      %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n =      %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n =      %d,  mu_n = %.11f " %(n,ans))

#pi approx for n = ONE HUNDRED
n = 100
ans = rho_pi(n)
print ("When n =     %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n =     %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n =     %d,  mu_n = %.11f " %(n,ans))

#pi approx for n = ONE THOUSAND
n = 1000
ans = rho_pi(n)
print ("When n =    %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n =    %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n =    %d,  mu_n = %.11f " %(n,ans))

#pi approx for n = TEN THOUSAND
n = 10000
ans = rho_pi(n)
print ("When n =   %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n =   %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n =   %d,  mu_n = %.11f " %(n,ans))

#pi approx for n = ONE HUNDRED THOUSAND
n = 100000
ans = rho_pi(n)
print ("When n =  %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n =  %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n =  %d,  mu_n = %.11f " %(n,ans))


#pi approx for n = ONE MILLION DOLLARS!! (https://www.youtube.com/watch?v=l91ISfcuzDw)
#
#n = ONE MILLION
n = 1000000
ans = rho_pi(n)
print ("When n = %d, rho_n = %.11f " %(n,ans))

ans = tau_pi(n)
print ("When n = %d, tau_n = %.11f " %(n,ans))

ans = mu_pi(n)
print ("When n = %d,  mu_n = %.11f " %(n,ans))


```
